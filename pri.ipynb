```python
# FILTRAR DATOS
# filtra una fila donde la clase es mayor igual a 2000
datos[(datos['CLASE'] >= 2000)]
```

```python
# Filtrar datos 
# filtra datos por una condicion multiple
# en este caso se flitraron las clases menores iguales a 2000 y se pide que el genero sea igual a masculino
datos[(datos['CLASE'] <=2000) & (datos['GENERO'] == 'M')]
```

```python
# Estamos filtrando datos especifico como los  de la columna 'TIPO DE EJEMPLAR'
datos[datos['TIPO DE EJEMPLAR'].isin(['DNI', 'DNI-EB'])]
```
