```python
#obtenemos el año actual
año_actual = datetime.now().year
```

```python
conteo_genero = df['GENERO'].value_counts()
print(conteo_genero)
```

```python
edad_promedio = df.groupby('TIPO DE EJEMPLAR')['EDAD'].mean()
print(edad_promedio)
```

```python
seccion_conteo = df['SECCION'].value_counts()
print(seccion_conteo)
```

```python
seccion_genero = df.groupby('SECCION')['GENERO'].value_counts()
print(seccion_genero)
```

```python
distribucion_edad = df['EDAD'].value_counts()
print(distribucion_edad)
```

```python
mayores_30 = df[df['EDAD'] > 30]
print(mayores_30)
```

```python
tabla_resumen = df.pivot_table(index='TIPO DE EJEMPLAR', columns='GENERO', values='EDAD', aggfunc='mean')
print(tabla_resumen)
```

```python
df['SECCION'].value_counts().plot(kind='bar', title='Conteo de Secciones')
```
