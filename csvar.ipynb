```python
import pandas as info
```

```python
import matplotlib.pyplot as plt
```

```python
datos = info.read_csv("./IMDB-Movie-Data.csv")
```

```python
datos
```

```python
datos.columns
```

```python
datos.plot(x="Rating", y="Votes", kind="scatter", color="red")
plt.title("IMDB Movie Data")
```

```python
datos.plot(x="Year", y="Rating", kind="line", color="blue")
plt.title("IMDB Movie Data")
```

```python
datos.columns
```

```python
datos['Revenue (Millions)'].plot(kind='hist', color='green')
plt.title("IMDB Movie Data")
plt.xlabel("Revenue (Millions)")
```

```python
import seaborn as sns
sns.set(style="whitegrid")
```

```python
sns.boxplot(x="Year", y="Rating", data=datos)
plt.title("IMDB Movie Data")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x="Votes", y="Rating", hue="Year", data=datos)
plt.title("IMDB Movie Data")
```

```python
sns.barplot(x="Year", y="Rating", data=datos)
plt.title("IMDB Movie Data")
```

```python
sns.violinplot(x="Year", y="Rating", data=datos)
```

```python
sns.pairplot(datos, hue="Year", diag_kind="kde")
plt.title("IMDB Movie Data")
plt.show()
```

```python
sns.regplot(x="Votes", y="Rating", data=datos)
plt.title("IMDB Movie Data")
plt.show()
```
